```python
sales_copy = sales[1:, 1:]
sales_copy
```

```python
month_averages = []

for i in range(0,len(sales),4):
    month_sales = sales_copy[i:i+4]#slice by 4 week interval
    print(month_sales, "\n")
    month_mean = month_sales.mean()
    month_averages.append(month_mean) #append new average to the list of averges
    
print(month_averages)
highest_sales = max(month_averages) #get highest value in month list
print(highest_sales)
print(month_averages.index(highest_sales)) #get list position of highest sales value
    
    
```
